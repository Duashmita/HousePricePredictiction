# Calculating the loss and accuracy of the model

## Importing Libraries

```python
import matplotlib.pyplot as plt
from main import loss_train,loss_test
```

## Ploting the loss for training vs testing

```python
plt.figure(figsize=(10, 5))
plt.plot(loss_train, label='Train Loss', marker='o')
plt.plot(loss_test, label='Test Loss', marker='x')
plt.xlabel('Batch or Epoch')
plt.ylabel('Loss')
plt.title('Train and Test Loss Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
